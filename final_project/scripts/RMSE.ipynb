```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
import os
```

```python

# Specify the path to your folder containing CSV files
folder_path = '/content/drive/My Drive/afcs/forecasts/'

# Create an empty dictionary to store dataframes
dfs = {}

# Iterate through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        # Construct the full path to the CSV file
        file_path = os.path.join(folder_path, filename)

        # Read the CSV file into a dataframe and store it in the dictionary
        df = pd.read_csv(file_path, index_col=0)

        # Remove the ".csv" extension from the filename to use as the dictionary key
        key = os.path.splitext(filename)[0]

        # Store the dataframe in the dictionary with the filename as the key
        dfs[key] = df
```

```python
test_file = pd.read_csv('/content/drive/My Drive/afcs/data/sales_test_validation_afcs2022.csv')
```

```python
def get_rmse(forecasts, test):
  test = test.drop('id', axis=1)
  rmse = {}

  for key, df in dfs.items():
    df = df.iloc[:, 1:]
    df = df.fillna(0)
    rmse.update({key: mean_squared_error(test.values, df.values, squared=False)})

  rmse = pd.DataFrame.from_dict(rmse, orient="index", columns=["RMSE"]).sort_values('RMSE')

  return rmse

```

```python
rmse = get_rmse(dfs, test_file)
rmse
```
